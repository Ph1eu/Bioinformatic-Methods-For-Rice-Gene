```python
import pandas as pd
import seaborn as sns
```

<a name='2'></a>
# 2 - Grain Properties


```python
df=pd.read_csv("./data/PhenotypeData/GrainPropertiesInVietnam.csv")
df
```

<a name='3'></a>
# 3 - Grain weight


```python
sns.set_theme()
distweight = sns.displot(df, x="Grain weight",bins=7)
distweight.fig.set_size_inches(15,6)

```

<a name='4'></a>
# 4 - Grain length


```python
sns.set_theme()
distlength = sns.displot(df, x="grain length",
                         color="red",
                         bins=7
                              )
distlength.fig.set_size_inches(15,6)
```

<a name='5'></a>
# 5 - Grain width


```python
sns.set_theme()
distwidth = sns.displot(df, x="grain width",
                         bins=5,
                         color="yellow"
                              )
distwidth.fig.set_size_inches(15,6)
```
